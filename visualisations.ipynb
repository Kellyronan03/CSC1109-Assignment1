```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('Data/cleaned_dataset.csv')
```

```python
# Genres with highest average energy
avg_energy_by_genre = df.groupby('Genre')['Energy'].mean().reset_index()
avg_energy_by_genre.columns = ['Genre', 'Avg_Energy']
top_genres = avg_energy_by_genre.nlargest(10, 'Avg_Energy')

plt.figure(figsize=(12, 6), facecolor='white')
sns.barplot(data=top_genres, x='Genre', y='Avg_Energy')
plt.xlabel('Genre')
plt.ylabel('Average Energy')
plt.title('Top 10 Genres by Average Energy')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Genres with highest average danceability
avg_energy_by_genre = df.groupby('Genre')['Danceability'].mean().reset_index()
avg_energy_by_genre.columns = ['Genre', 'Avg_Danceability']
top_genres = avg_energy_by_genre.nlargest(10, 'Avg_Danceability')

plt.figure(figsize=(12, 6), facecolor='white')
sns.barplot(data=top_genres, x='Genre', y='Avg_Danceability')
plt.xlabel('Genre')
plt.ylabel('Average Danceabilityy')
plt.title('Top 10 Genres by Average Danceability')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Corellation between energy and danceability
import seaborn as sns
import matplotlib.pyplot as plt
#Sampling the data to improve readability
limited_data = df.sample(10000, random_state=1)

plt.figure(figsize=(12, 6), facecolor='white')
sns.regplot(data=limited_data, x='Danceability', y='Energy', scatter_kws={'alpha':0.6}, line_kws={'color': 'red', 'linestyle': '--'})
plt.title('Energy vs Danceability')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.savefig('../../energyVdanceability.png')
plt.grid()
plt.show()
```

```python
# Average Danceability by time signature
avg_danceability = df.groupby('Time Signature')['Danceability'].mean().reset_index() # Group time signatures by average danceability
avg_danceability = avg_danceability.sort_values(by='Danceability', ascending=False) # Sort the values
plt.figure(figsize=(12, 6), facecolor='white')
sns.barplot(data=avg_danceability, x='Time Signature', y='Danceability', palette='viridis')
plt.title('Average Danceability by Time Signature')
plt.xlabel('Time Signature')
plt.ylabel('Average Danceability')
plt.xticks(rotation=45) 
plt.grid(axis='y')
plt.savefig('../../Average_Danceability_by_Time_Signature.png')
plt.show()
```
